Here is the step-by-step explanation with the corresponding Python code:

1. Create a Simple Dataset:

```python
import numpy as np
import pandas as pd

# Define the features and the target variable.
data = {
    'Size':  [0, 0, 1, 1, 0],  # 0 for small, 1 for large
    'Color': [0, 0, 0, 1, 1],  # 0 for red, 1 for blue
    'Class': [0, 0, 0, 1, 1]   # 0 for Class A, 1 for Class B
}

df = pd.DataFrame(data)
```

This code block creates a DataFrame with two features (Size and Color) and a binary target (Class).

2. Split the Data into Features and Target:

```python
# Splitting the dataset into features (X) and target (y)
X = df[['Size', 'Color']]
y = df['Class']
```

Here, X contains the features, and y contains the target class labels.

3. Train a Decision Tree Classifier:

```python
from sklearn.tree import DecisionTreeClassifier

# Initialize the DecisionTreeClassifier with a max depth of 2 for simplicity
tree_clf = DecisionTreeClassifier(max_depth=2, random_state=42)

# Fit the model to the data
tree_clf.fit(X, y)
```

A DecisionTreeClassifier is instantiated and trained on the simple dataset. The max_depth parameter is set to 2 to keep the tree simple.

4. Plot the Decision Tree:

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# Plot the decision tree
plt.figure(figsize=(12, 8))
plot_tree(tree_clf, filled=True, feature_names=['Size', 'Color'], class_names=['Class A', 'Class B'])
plt.show()
```
